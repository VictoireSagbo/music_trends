```python
# Importing necessary libraries for data collection, manipulation and visualization

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import matplotlib.pyplot as plt
```

## II- Collecting Data

**1. API authentication**

```python
# Spotify API authentication

client_id = 'Your_client_id'
client_secret = 'Your_client_secret'
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

**2. Retrieving playlist**

```python
# Function for retrieving playlist song data
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

**3. Retrieving popular playlists song data**

We will be retrieving the playlist from US.

```python
# Function for retrieving popular playlist song data

def get_category_playlists(category_id, country='US', limit=10):
    playlists = sp.category_playlists(category_id=category_id, country=country, limit=limit)
    return playlists['playlists']['items']
```

**4. Extracting relevant information from songs**

```python
# Extract relevant information from songs
def extract_track_info(tracks):
    track_data = []
    for track in tracks:
        track_info = track['track']
        track_data.append({
            'name': track_info['name'],
            'artist': track_info['artists'][0]['name'],
            'album': track_info['album']['name'],
            'release_date': track_info['album']['release_date'],
            'duration_ms': track_info['duration_ms'],
            'popularity': track_info['popularity']
        })
    return track_data
```

**5. Retrieving popular playlists in the a specific category**

```python
# Retrieve popular playlists in the "pop" category
pop_playlists = get_category_playlists('pop')

# Extract song data from these playlists
all_tracks = []
for playlist in pop_playlists:
    tracks = get_playlist_tracks('spotify', playlist['id'])
    track_data = extract_track_info(tracks)
    all_tracks.extend(track_data)
```

###  III.  Data Cleaning and Transformation

```python
# Create a pandas DataFrame
df = pd.DataFrame(all_tracks)

# Convert song duration from milliseconds to minutes
df['duration_min'] = df['duration_ms'] / 60000

# Display the first lines of the DataFrame
df.head()
```

### V. Data analysis and visualization

1. **Top Artists**:


```python
# Histogram of musical genres

plt.figure(figsize=(12, 6))
df['artist'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Most Frequent Artists in Popular Playlists')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.show()
```

**Comments:**
***This bar chart shows the top 10 artists who appear most frequently in popular playlists.
T***

***The result shows that the most popular is Taylor Swift with more than 20 songs***

2. **Song Duration Distribution**:


```python

# Average song length
plt.figure(figsize=(12, 6))
plt.hist(df['duration_min'], bins=30, edgecolor='k')
plt.title('Distribution of Song Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Songs')
plt.show()
```

**Comments:**
This histogram shows the distribution of song durations in the dataset.
The x-axis represents the duration of songs in minutes, while the y-axis represents the number of songs.

***Most songs have durations clustering around 3 to 4 minute.
This clustering indicates that 3 to 4 minutes is a common duration for songs in the dataset, which is typical for many popular music genres.***

**3- Popularity vs. Duration**:


```python

# Popularity of songs
plt.figure(figsize=(12, 6))
plt.scatter(df['duration_min'], df['popularity'], alpha=0.5)
plt.title('Popularit des morceaux en fonction de leur dure')
plt.title('Popularity of Songs Based on Their Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Popularity')
plt.show()
```

**Comments:**
   - Most popular songs (popularity scores between 60 to 100) are clustered between 2.5 and 4.5 minutes in duration.
